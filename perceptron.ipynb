```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 1: Or Problem

## Online Method

```python
input = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
target = np.array([0, 1, 1, 1]) # y
b = random.random() # bias
wt = np.array([random.random() for _ in range(2)]) # weights
alpha = 0.15 # learning rate
errors = []
epoch = 1
costs = 1

print(wt)
while costs / 4 > 0.002:
    costs = 0
    print(f"Epoch {epoch}:")
    for i in range(4):
        output = (wt.T @ input[i]) - b
        phi = 1 if output >= 0 else 0
        error = target[i] - phi
        cost = (1 / (2 * 4)) * np.sum(np.square(error))
        wt = wt + (alpha * error * input[i])
        b = b - (alpha * error) # Update bias

        print(f"Target: {target[i]} | Error: {cost:.3f} | Phi: {phi}")
        costs += cost

    errors += [costs / 4]
    epoch += 1
    print()

if len(errors) > 1:
    plt.plot(errors)
    plt.show()
```

## Offline Method

```python
input = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
target = np.array([0, 1, 1, 1]) # y
b = random.random() # bias
wt = np.array([random.random() for _ in range(2)]) # weights
alpha = 0.15 # learning rate
errors = []
cost = 10
epoch = 1

while cost > 0.01:
    print(f"Epoch {epoch}:")
    output = input @ wt.T + b
    phi = [1 if element >= 0 else 0 for element in output]
    error = target - phi
    cost = (1 / (2 * 4)) * np.sum(np.square(error))

    print(f"Target: {target} | phi: {phi}")
    errors += [cost]

    wt = wt + (alpha * (error @ input))
    b = b + (alpha * np.sum(error)) # Update bias

    epoch += 1
    print()

if len(errors) > 1:
    plt.plot(errors)
    plt.show()
```

# Exercise 2: 2 Perceptrons
1 produces or, the other produces nor

## 2 Online Method

```python
input = np.array([
        [1, 1],
        [0, 1],
        [1, 0],
        [0, 0]
])
target_or = np.array([1, 1, 1, 0])
target_nor = np.array([0, 0, 0, 1])
wt = np.array([[random.random(), random.random()] for _ in range(2)])
alpha = 0.15 # learning rate
a = 1
b1 = 1
b2 = 1
epoch = 1
errors = []
costs = 10

while costs / 4 > 0.001:
    costs = 0
    print(f"Epoch {epoch}:")
    for i in range(4):
        v1 = wt[0].T @ input[i] - b1
        v2 = wt[1].T @ input[i] - b2

        y1 = 1 / (1 + np.exp(-a * v1))
        y2 = 1 / (1 + np.exp(-a * v2))

        error1 = target_or[i] - y1
        error2 = target_nor[i] - y2

        delta1 = error1 * y1 * (1 - y1)
        delta2 = error2 * y2 * (1 - y2)

        wt[0] = wt[0] + alpha * delta1 * input[i]
        wt[1] = wt[1] + alpha * delta2 * input[i]

        b1 = b1 - alpha * delta1
        b2 = b2 - alpha * delta2

        cost = (1 / (2 * 4)) * (np.square(error1) + np.square(error2))

        print(f"Target 1: {target_or[i]} | Target 2: {target_nor[i]} | y1: {y1:.3f} | y2: {y2:.3f}")
        costs += cost

    errors.append(costs / 4)
    epoch += 1
    print()

if len(errors) > 1:
    plt.plot(errors)
    plt.show()
```
